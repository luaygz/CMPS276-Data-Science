## Some helper methods

```python
class color:
    BOLD = '\033[1m'
    END = '\033[0m'

def print_bold(text):
    print(color.BOLD + text + color.END)
```

```python
def summarize_numeric_categorical(col):
    print_bold(f"-{col.name}-\n")
    print(f"Number unique: {col.nunique()}\n")
    print(f"Unique values: {sorted(col.unique())}\n")
    print(f"Unique counts: \n{col.value_counts()}\n")
    print(f"Unique counts by percent: \n{col.value_counts()/col.size*100}\n")
    mean_std_quantiles = col.describe().apply(lambda s: "{0:.2f}".format(s)) # Two decimal places
    print(f"Metrics: \n{mean_std_quantiles}\n")
    col.value_counts().sort_index().plot(kind="bar")
```

```python
def summarize_numeric(col):
    print_bold(f"-{col.name}-\n")
    mean_std_quantiles = col.describe().apply(lambda s: "{0:.2f}".format(s)) # Two decimal places
    print(f"Metrics: \n{mean_std_quantiles}\n")
    col.plot.box()
```

```python
def summarize_categorical(col):
    print_bold(f"-{col.name}-\n")
    print(f"Number unique: {col.nunique()}\n")
    print(f"Unique values: {sorted(col.unique())}\n")
    print(f"Unique counts: \n{col.value_counts()}\n")
    print(f"Unique counts by percent: \n{col.value_counts()/col.size*100}\n")
    col.value_counts().sort_index().plot(kind="bar")
```

# Let's explore how different factors impact severity

### Some helper functions

```python
def discretize_and_aggregate_mean(df, var, agg_by, bin_size=1.0):
    _1st_percentile = df[agg_by].quantile(0.01)
    _99th_percentile = df[agg_by].quantile(0.99)
    discretized = df[agg_by].apply(lambda x: round(x/bin_size, 0)*bin_size)
    mean_of_var_vs_agg = df.groupby(discretized).mean()[var]
    plt.plot(mean_of_var_vs_agg)
    plt.axvline(x=_1st_percentile)
    plt.axvline(x=_99th_percentile)
    plt.xlabel(agg_by, fontsize=20)
    plt.ylabel(f"Average {var}", fontsize=20)
    plt.show()

def categorical_aggregate_mean(df, var, agg_by, sort_by_value=False):
    mean_of_var_vs_agg = df.groupby(agg_by).mean()[var]
    low = min(mean_of_var_vs_agg)
    high = max(mean_of_var_vs_agg)
    diff = high - low
    plt.ylim([low - diff/2, high+diff/2])
    plt.xticks(rotation="vertical")
    plt.xlabel(agg_by, fontsize=20)
    plt.ylabel(f"Average {var}", fontsize=20)
    if sort_by_value:
        mean_of_var_vs_agg.sort_values().plot.bar(width=0.8)
    else:
        mean_of_var_vs_agg.plot.bar(width=0.8)
    plt.show()
```

## The data becomes very sparse at the lower and upper extremes, so we mark the 1st and 99th percentiles with vertical lines beyond which the data may be unrepresentative.

## How does distance impact severity?

```python
discretize_and_aggregate_mean(df, "Severity", "Distance(mi)")
```

### We can see a positive approximately linear relationship between severity and distance, where severity ranges between around 2.2 and 2.6

## How does temperature impact severity?

```python
discretize_and_aggregate_mean(df, "Severity", "Temperature(F)")
```

### There seems to be a small dip in the middle. We can see that the severity of accidents is lowest at ~55 and increases as the temperature gets colder or warmer. The mean severity seems to range beteen 2.2 and 2.35

## How does wind chill impact severity?

```python
discretize_and_aggregate_mean(df, "Severity", "Wind_Chill(F)")
```

### Again we see a dip around the middle and an increase of severity as one moves away from the center. The mean severity seems to range beteen 2.2 and 2.35

## How does humidity impact severity?

```python
discretize_and_aggregate_mean(df, "Severity", "Humidity(%)")
```

### It seems the severity of accidents is low when it is dry, increases approximately linearly until around 40% humidity, then it levels off, only to start increasing again around 90% humidity. The mean severity seems to range beteen 2.0 and 2.5

## How does atmospheric pressure impact severity?

```python
discretize_and_aggregate_mean(df, "Severity", "Pressure(in)")
```

### We can see a positive approximately linear relationship between severity and pressure ranging between 1.95 and 2.35

```python
discretize_and_aggregate_mean(df, "Severity", "Precipitation(in)", bin_size=0.01)
```

### There seems to be a sharp increase then it gets lower. Let's make better sense of this. A precipitation of zero means no rain, and a precipitation of any positive number signifies that it is raining.
