```python
import pandas as pd
from sklearn.cluster import KMeans
import plotly.express as px
```

```python
df = pd.read_csv("US_Accidents_Dec20_updated_cleaned_imputed.csv")
```

### These are strings and can't be clustered

```python
df.drop(["Start_Time", "End_Time", "State", "Wind_Direction", "Weather_Condition"], axis = 1, inplace = True)
```

```python
kmeans_3_clusters = KMeans(n_clusters=3, random_state=0, verbose=1)
kmeans_3_clusters.fit(df)
```

```python
centroids = kmeans_3_clusters.cluster_centers_
print(centroids)
```

```python
n = 500000 # It doesn't run if we use the whole dataset
fig = px.scatter_3d(df.iloc[:n], x="Humidity(%)", y="Precipitation(in)", z="Temperature(F)",
                    color=kmeans_3_clusters.labels_.astype(float)[:n])#, symbol="species")
fig.show()
```

```python
kmeans_6_clusters = KMeans(n_clusters=6, random_state=0, verbose=1)
kmeans_6_clusters.fit(df)
```

```python
centroids = kmeans_6_clusters.cluster_centers_
print(centroids)
```

```python
n = 500000 # It doesn't run if we use the whole dataset
fig = px.scatter_3d(df.iloc[:n], x="Humidity(%)", y="Precipitation(in)", z="Temperature(F)",
                    color=kmeans_6_clusters.labels_.astype(float)[:n])#, symbol="species")
fig.show()
```
